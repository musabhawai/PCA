```python
from sklearn.datasets import load_digits
import pandas as pd

dataset = load_digits()
dataset.keys()
```

```python
dataset.data.shape
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
dataset.target
```

```python
df.describe()
```

```python
X = df
y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
#Use PCA to reduce dimensions
```

```python
X
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_ # each column here represent a percent of total info like 1 0.1490594 means 14% of over all information
```

```python
pca.n_components_ #PCA created 29 components out of 64 original columns
```

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
#Let's now select only two components
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_ #You can see that both combined retains 0.14+0.13=0.27 or 27% of important feature information
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
#We get less accuancy (~60%) as using only 2 components did not retain much of the feature information. 
#SHowever in real life you will find many cases where using 2 or few PCA components can still give you a pretty good accuracy
```
